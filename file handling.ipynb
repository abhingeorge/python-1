```python
#python program to cody odd lines of one file to other
file1=input("Enter the source file")
file2=input("Enter the destination file")
f1=open(file1,"r")
f2=open(file2,"w")
i=0
for line in f1.readlines():
    i=i+1
    if i%2==0:
        continue
    f2.write(line)
f1.close()
f2.close()
```

```python
# 4. Write a Python program to read specific columns of a given CSV file and print the content of the columns. 
import csv
filename = "student.csv"
with open(filename, 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        print(row['Std Names'], "-", row['Std Total Marks'])
```

```python
 #5. Write a Python program to write a Python dictionary to a csv file. 
# After writing the CSV file read the CSV file and display the content.
import csv
students = [
    {"Name": "Alice", "Age": 22, "City": "Chennai"},
    {"Name": "Bob", "Age": 25, "City": "Delhi"},
    {"Name": "Carol", "Age": 23, "City": "Mumbai"}
]
filename = "students.csv"
# Writing dictionary to CSV
with open(filename, 'w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=["Name", "Age", "City"])
    writer.writeheader()
    writer.writerows(students)
print("Data written successfully to students.csv")
# Reading the same file
with open(filename, 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```
