```python
#1.Create a class Rectangle with private attributes length and width. Overload ‘<’ operator to compare the area of 2 rectangles. 
class Rectangle:
    def __init__(self, l, b):
        self.length = l
        self.breadth = b
        self.area = l * b

    def __lt__(self, other):
        if self.area < other.area:
            print("Rectangle 2 is greater")
        else:
            print("Rectangle 1 is greater")

l1 = int(input("Enter length of rectangle 1: "))
w1 = int(input("Enter width of rectangle 1: "))
l2 = int(input("Enter length of rectangle 2: "))
w2 = int(input("Enter width of rectangle 2: "))

r1 = Rectangle(l1, w1)
r2 = Rectangle(l2, w2)
r1 < r2
```

```python
# 2.create a class time with privateattributes hour,minute and second,overload '+' operator to find sum of two time.
class time:
    def __init__(self,hour,minute,second):
        self.hour=hour
        self.minute=minute
        self.second=second
    def display(self):
        print(self.hour,":",self.minute,":",self.second)
    
    def __add__(self,new):
        h=self.hour+new.hour
        m=self.minute+new.minute
        s=self.second+new.second
        if s>=60:
            m+=s//60
            s=s%60
        if m>=60:
            h+=m//60
            m=m%60
        return time(h,m,s)            
th1=int(input("enter the hour of first time:"))
tm1=int(input("enter the minute of first time:"))
ts1=int(input("enter the second of first time:")) 
th2=int(input("enter the hour of 2nd time:"))
tm2=int(input("enter the minute of 2nd time:"))
ts2=int(input("enter the second of 2nd time:")) 
print("first time=")
t1=time(th1,tm1,ts1)
t1.display()
print("2nd time=")
t2=time(th2,tm2,ts2)
t2.display()
print("sum of the time=")
t3=t1+t2
t3.display()
```
